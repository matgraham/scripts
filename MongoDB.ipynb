```python
db = client.business
```

```python
names = ['Kitchen','Animal','State','Tasty','Big','City','Fish','Pizza','Coffee']
```

```python
company_type = ['LLC','Inc', 'Company', 'Corporation']
```

```python
company_cuisine =['Pizza','Bar Food','Fast Food','Italian','Mexican']
```

```python
for x in range(1,501):
    business = {
        'name' : names[randint(0, (len(names)-1))] + ' ' + names[randint(0,(len(names)-1))] + ' ' + company_type[randint(0, (len(company_type)- 1))],
        'rating' : randint(1,5),
        'cuisine' : company_cuisine[randint(0, (len(company_cuisine)-1))]
    }
    result = db.reviews.insert_one(business)
    print('Created {0} of 500 as {1}'.format(x,result.inserted_id))
```

```python
fivestar = db.reviews.find_one({'rating':5})
print(fivestar)
```

```python
fivestarcount = db.reviews.find({'rating':5}).count()
print(fivestarcount)
```

```python
print('The sum of each rating occurance accross all data grouped by rating ')
stargroup = db.reviews.aggregate(
[
{ '$group':
     { '_id': '$rating',
     'count':
         {'$sum':1}
     }
},
{'$sort': {'_id':1}
}
])
for group in stargroup:
    print(group)
```
